```python
import pandas as pd
import numpy as np
```

```python
dd =pd.date_range(start='1/1/2020', end='31/12/2020')
```

```python
dd_array = np.array(dd,dtype= 'str')
```

```python
datessss = []
for i in np.arange(0,len(dd_array)):
    year  = dd_array[i][0:4]
    month = dd_array[i][5:7]
    day   = dd_array[i][8:10]
    one = year+month+day
    datessss.append(one)
dates = np.array(datessss,dtype='str')
```
